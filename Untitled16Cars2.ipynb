```python
import yaml

with open('/content/dataset/data.yaml') as f:
    data = yaml.safe_load(f)

print(data)
```

```python
import os
from collections import Counter

label_dir = '/content/dataset/train/labels'
class_counter = Counter()

for file_name in os.listdir(label_dir):
    with open(os.path.join(label_dir, file_name)) as f:
        for line in f:
            cls = int(line.split()[0])
            class_counter[cls] += 1

# Сортировка по номеру класса
for cls in sorted(class_counter):
    print(f"Класс {cls} ({data['names'][cls]}): {class_counter[cls]} объектов")
```

```python
import matplotlib.pyplot as plt

# Классы и их названия
classes = [data['names'][i] for i in sorted(class_counter)]
counts = [class_counter[i] for i in sorted(class_counter)]

plt.figure(figsize=(10,6))
plt.bar(classes, counts, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Классы повреждений')
plt.ylabel('Количество объектов')
plt.title('Распределение объектов по классам в тренировочном наборе')
plt.tight_layout()
plt.show()
```

```python
def get_augmented(image, bboxes, class_labels):
    transform = A.Compose([
        A.RandomBrightnessContrast(p=0.5),
        A.HorizontalFlip(p=0.5),
        A.RandomRain(slant_lower=-10, slant_upper=10, drop_length=20, drop_width=1, p=0.5),  # дождь
        A.RandomFog(fog_coef_lower=0.1, fog_coef_upper=0.3, p=0.3),  # туман / пыль
        A.GaussNoise(var_limit=(10.0, 50.0), p=0.2),
    ], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))

    augmented = transform(image=image, bboxes=bboxes, class_labels=class_labels)
    return augmented['image'], augmented['bboxes'], augmented['class_labels']
```

```python
import albumentations as A
import numpy as np

# Количество объектов по классам
class_counts = np.array([100,207,1,170,135,125,41,173,398,552,348,709,32,292,375,161])
rare_classes = [i for i, c in enumerate(class_counts) if c < 100]

def augment_rare(image, bboxes, class_labels):
    # Применяем аугментацию только если есть редкие классы
    if any(cl in rare_classes for cl in class_labels):
        transform = A.Compose([
            A.RandomBrightnessContrast(p=0.5),
            A.HorizontalFlip(p=0.5),
            A.Rotate(limit=15, p=0.5),
            A.GaussNoise(var_limit=(10.0,50.0), p=0.3)
        ], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels']))
        augmented = transform(image=image, bboxes=bboxes, class_labels=class_labels)
        return augmented['image'], augmented['bboxes'], augmented['class_labels']
    else:
        return image, bboxes, class_labels
```

```python
from ultralytics import YOLO

# Загружаем предобученную модель YOLOv8s на COCO
model = YOLO("yolov8s.pt")  # small, можно заменить на yolov8m/l для большей точности

# Путь к data.yaml
data_yaml = "/content/dataset/data.yaml"

# Запуск тренировки
model.train(
    data=data_yaml,
    epochs=80,
    batch=16,
    imgsz=640,  # можно увеличить до 768 для мелких царапин
    lr0=0.003,  # пониженный lr для стабильного обучения
    device=0,
    optimizer="SGD",
    augment=True,  # встроенная YOLO аугментация
    name="yolov8s_car_dents_balanced",
    project="/content/runs/detect"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
dataset_path='/content/dataset'
results = model.predict(
    source=os.path.join(dataset_path, "test/images"),
    conf=0.25,
    save=True
)
```

```python
from ultralytics import YOLO
import os

metrics = model.val()

# Средние метрики по всем классам
precision = metrics.box.map.mean()  # mAP по IoU=0.5
recall = metrics.box.r.mean()       # recall
f1 = metrics.box.f1.mean()          # F1-score

print(f"Средние метрики по всем классам:")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")

# Если нужно, метрики по каждому классу
print("\nМетрики по классам:")
for cls, p, r, f in zip(metrics.names.values(), metrics.box.p, metrics.box.r, metrics.box.f1):
    print(f"{cls}: Precision={p:.3f}, Recall={r:.3f}, F1={f:.3f}")
```
